```python
def plot_vectors(vectors, colors):
    """
    Plot one or more vectors in a 2D plane, specifying a color for each.

    Arguments
    ---------
    vectors: list of lists or of arrays
        Coordinates of the vectors to plot. For example, [[1, 3], [2, 2]]
        contains two vectors to plot, [1, 3] and [2, 2].
    colors: list
        Colors of the vectors. For instance: ['red', 'blue'] will display the
        first vector in red and the second in blue.

    Example
    -------
    plot_vectors([[1, 3], [2, 2]], ['red', 'blue'])
    plt.xlim(-1, 4)
    plt.ylim(-1, 4)
    """
    plt.figure()
    plt.axvline(x=0, color='lightgray')
    plt.axhline(y=0, color='lightgray')

    for i in range(len(vectors)):
        x = np.concatenate([[0,0],vectors[i]])
        plt.quiver([x[0]], [x[1]], [x[2]], [x[3]],
                   angles='xy', scale_units='xy', scale=1, color=colors[i],)
```

**Regression in PyTorch**

use the PyTorch automatic differentiation library to fit a straight line to data points.

```python
#E.g.: Dosage of drug for treating Alzheimer's disease
x = torch.tensor([0, 1, 2, 3, 4, 5, 6, 7.])
x
```

```python
y = torch.tensor([1.86, 1.31, .62, .33, .09, -.67, -1.23, -1.37]) # E.g.: Patient's "forgetfulness score"
y
```

```python
fig, ax = plt.subplots()
plt.title("Clinical Trial")
plt.xlabel("Drug dosage (mL)")
plt.ylabel("Forgetfulness")
ax.scatter(x, y)
```

```python
m = torch.tensor([0.9]).requires_grad_()
m
```

```python
b = torch.tensor([0.1]).requires_grad_()
b
```

```python
def regression(my_x, my_m, my_b):
    return my_m*my_x + my_b
```

```python
def regression_plot(my_x, my_y, my_m, my_b):

    fig, ax = plt.subplots()

    ax.scatter(my_x, my_y)

    x_min, x_max = ax.get_xlim()
    y_min = regression(x_min, my_m, my_b).detach().item()
    y_max = regression(x_max, my_m, my_b).detach().item()

    ax.set_xlim([x_min, x_max])
    _ = ax.plot([x_min, x_max], [y_min, y_max])
```

```python
regression_plot(x, y, m, b)
```

**Machine Learning with Automatic Differentiation**



*   Step 1. Forward pass
*   Step 2. Compare yhat with true y to calculate MSE Cost(Loss) C
*   Step 3. Use Chain Rule to calculate gradient of C w.r.t. parameters m and b
*   Step 4. Adjust m and b to reduce C




```python
yhat = regression(x, m, b)
yhat
```

```python
# create our own MSE function to calculate the cost
def mse(my_yhat, my_y):
  sigma = torch.sum((my_yhat - my_y)**2)
  return sigma/len(my_y)
```

```python
C = mse(yhat, y)
C
```

```python
C.backward()
```

```python
m.grad
```

```python
b.grad
```

```python
# gradient descent
optimizer = torch.optim.SGD([m, b], lr=0.01)
```

```python
optimizer.step()
```

```python
# now, confirm parameters have been adjusted sensibly:
m
```

```python
b
```

```python

regression_plot(x, y, m, b)
```

```python
# conform cost has decreased
C = mse(regression(x, m, b), y)
C
```

```python
# Put the 4 steps in a loop to iteratively minimize cost toward zero:
epochs = 30
for epoch in range(epochs):
  optimizer.zero_grad()
  yhat = regression(x, m, b)
  C = mse(yhat, y)
  C.backward()
  optimizer.step()
  print('Epoch {}, cost {}, m grad {}, b grad {}'.format(epoch, '%.3g' % C.item(), '%.3g' % m.grad.item(), '%.3g' % b.grad.item()))
```

```python
regression_plot(x, y, m, b)
```

```python
epochs = 300
for epoch in range(epochs):
  optimizer.zero_grad()
  yhat = regression(x, m, b)
  C = mse(yhat, y)
  C.backward()
  optimizer.step()
  print('Epoch {}, cost {}, m grad {}, b grad {}'.format(epoch, '%.3g' % C.item(), '%.3g' % m.grad.item(), '%.3g' % b.grad.item()))
```

```python
regression_plot(x, y, m, b)
```

**Partial Derivatives of Multivariate Functions**

```python
def f(my_x, my_y):
  return my_x**2 - my_y**2
```

```python
xs = np.linspace(-3, 3, 1000)
```

```python
zs_wrt_x = f(xs, 0)
```

```python
fig, ax = plt.subplots()

plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

plt.xlabel('x')
plt.ylabel('z', rotation=0)

ax.plot(xs, zs_wrt_x)
```

```python
# to determine the slope of z wrt x at any given point, we can use the partial derivative:
def delz_delx(my_x, my_y):
  return 2*my_x
```

```python
x_samples = [-2, -1, 0, 1, 2]
```

```python
colors = ['red', 'orange', 'green', 'blue', 'purple']
```

```python
def point_and_tangent_wrt_x(my_xs, my_x, my_y, my_f, fprime, col):

    my_z = my_f(my_x, my_y) # z = f(x, y)
    plt.scatter(my_x, my_z, c=col, zorder=3)

    tangent_m = fprime(my_x, my_y) # Slope is partial derivative of f(x, y) w.r.t. x
    tangent_b = my_z - tangent_m*my_x # Line is z=mx+b, so b=z-mx
    tangent_line = tangent_m*my_xs + tangent_b

    plt.plot(my_xs, tangent_line, c=col,
             linestyle='dashed', linewidth=0.7, zorder=3)
```

```python
fig, ax = plt.subplots()
plt.axvline(x=0, color='lightgray')
plt.axhline(y=0, color='lightgray')

for i, x in enumerate(x_samples):
    point_and_tangent_wrt_x(xs, x, 0, f, delz_delx, colors[i])

plt.ylim(-1, 9)
plt.xlabel('x')
plt.ylabel('z', rotation=0)
_ = ax.plot(xs, zs_wrt_x)
```

**Greadient of a Single-Point Regression**

```python
xs = torch.tensor([0, 1, 2, 3, 4, 5, 6, 7.])
ys = torch.tensor([1.86, 1.31, .62, .33, .09, -.67, -1.23, -1.37])
```

```python
def regression(my_x, my_m, my_b):
  return my_m * my_x + my_b
```
